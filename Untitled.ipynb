```python
class Pile:
    def __init__(self):
        self.pile = []
    
    def __len__(self):
        return len(self.pile)
    
    def sommet(self):
        if len(self) > 0:
            return self.pile[0]
    
    def empiler(self, element):
        self.pile = [element] + self.pile
    
    def depiler(self):
        if len(self) > 0:
            return self.pile.pop(0)
    
class RPN:
    def __init__(self, calcul):
        self.calcul = calcul
    
    def addition(self):
        a = self.depiler()
        b = self.depiler()
        self.empiler(a + b)
    
    def soustraction(P):
        a = self.depiler()
        b = self.depiler()
        self.empiler(b - a)
    
    def multiplication(P):
        a = self.depiler()
        b = self.depiler()
        self.empiler(a*b)
    
    def division(P):
        a = self.depiler()
        b = self.depiler()
        self.empiler(b / a)

    
    def calculer(self):
        vrai = True
        nombre = ""
        rpn = Pile()
        while vrai:
            
            if self.calcul[0] in '0123456789':
                nombre += self.calcul[0]
            elif len(nombre) > 0:
                rpn.empiler(int(nombre))
                nombre = ""
            
            if self.calcul[0] == "+":
                rpn.addition()
            elif self.calcul[0] == "-":
                rpn.soustraction()
            elif self.calcul[0] == "/":
                rpn.division()
            elif self.calcul[0] == "*":
                rpn.multiplication()
            
            if len(self.calcul) > 0:
                self.calcul = self.calcul[1:]
            elif int(nombre) > 0 and len(self.calcul) == 0:
                rpn.empiler(nombre)
                vrai = False
            elif len(self.calcul) == 0:
                vrai = False
        return rpn.depiler()
```

```python
rpn = RPN("7 2 +")
rpn.calculer()
```
