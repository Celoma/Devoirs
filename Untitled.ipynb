```python
class Equation:
    """Classe permettant de résoudre des équations. Du degré 0 au deuxième degré
    """
    
    def __init__(self, list):
        
        signe = []
        nb = []
        
        for i in list:
            if i > 0:
                signe.append(1)
            else:
                signe.append(-1)

            nb.append(abs(i))
            
        self.signe = signe
        self.nb = nb
        
        
    def resoudre(self, rep = 0):

        if len(self.nb) == 1:
            return "Une constante n'a pas d'inconnu"
        
        elif len(self.nb) == 2:
            self.solution = [(- self.nb[1]*self.signe[1] + rep)/self.nb[0]]
        
        elif len(self.nb) == 3:
            
            delta = self.nb[1]**2 - (4*self.nb[0]* (self.nb[2] + rep))*self.signe[0]*self.signe[2]
            
            if delta < 0:
                return "Il n'y a pas de solution dans R"
            
            elif delta == 0:
                self.solition = [self.nb[1]*self.signe[1] / (2* self.nb[0]*self.signe[0])]
            
            else:
                Sa = (self.nb[1]*self.signe[1] + delta**0.5)  / (2* self.nb[0]*self.signe[0])
                Sb = (self.nb[1]*self.signe[1] - delta**0.5)  / (2* self.nb[0]*self.signe[0])
                self.solution = [Sa, Sb]
                
        elif len(self.nb) > 3:
            return f"L'équation n'est pas du second degré"
        
    def derive(self):
        for i in range(len(self.nb) - 2):
            print(f"{self.nb[i] * (len(self.nb)- i - 1) * self.signe[i]}x**{(len(self.nb)- i - 2)}", end = ' + ')
        print(f"{self.nb[i + 1] * (len(self.nb) - i - 2) * self.signe[i + 1]}")
    def __str__(self):
        
        for i in range(len(self.nb) - 1):
            print(f" {self.nb[i] * self.signe[i]}x**{(len(self.nb) - i - 1)}" , end = ' + ')
        print(f"{self.nb[len(self.nb)-1] * self.signe[len(self.signe)-1]}")
        
    def __repr__(self):

        for i in range(len(self.nb) - 1):
            print(f" {self.nb[i] * self.signe[i]}x**{(len(self.nb) - i - 1)}" , end = ' + ')
        print(f"{self.nb[len(self.nb)-1] * self.signe[len(self.signe)-1]}")
        
        
```
