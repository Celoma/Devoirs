```python
class Epee():
    def __init__(self, modele, force):
        self.modele = modele
        self.force = force
        self.est_utilise = False

    def combat(self, une_autre_epee):
        if self.force > une_autre_epee.force:
            une_autre_epee.casser()
            print("Vous avez gagner le duel")
            
        else :
            print("Vous avez perdu le duel")
            self.casser()

    def utiliser(self):
        self.est_utilise = True

    def ranger(self):
        self.est_utilise = False
    
    def casser(self):
        self.force = 0
```

```python
class Vecteur():
    """Class Vecteur
    
    self = Vecteur(x, y) 
    """
    
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __mul__(self, k):
        return Vecteur(k * self.x, k * self.y)
    
    def __add__(self, u):
        return Vecteur(u.x + self.x, u.y + self.y)
    
    def __sub__(self, u):
        return Vecteur(u.x - self.x, u.y - self.y)
    
    def __rmul__(self, k):
        return self.__mul__(k)
    
    def __str__(self):
        return f"({self.x}, {self.y})"
    
    def __repr__(self):
        return "Vecteur" + self.__str__()
```
