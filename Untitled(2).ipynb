```python
class Equation:
    """Classe permettant de résoudre des équations. Du degré 0 au deuxième degré
    """
    
    def __init__(self, list):
        
        signe = []
        nb = []
        
        for i in list:
            if i > 0:
                signe.append(1)
            else:
                signe.append(-1)

            nb.append(abs(i))
            
        self.signe = signe
        self.nb = nb
        
    def resoudre(self, rep = 0):
        
        if len(self.nb) > 3:
            return f"Cette équation n'a pas de solution pour {rep}"
        
        elif len(self.nb) == 1:
            return "Il n'y a pas d'inconnu"
        
        elif len(self.nb) == 3:
            
            delta = self.nb[1]**2 - (4*self.nb[0]* (self.nb[2] + rep))
            
            if delta < 0:
                return("Il n'y a pas de solution dans R")
            
            elif delta == 0:
                S = self.nb[1] / (2* self.nb[0])
                return(f"Pour obtenir {rep} avec l'équation x doit être égale à {S}")
            
            else:
                Sa = (self.nb[1] + delta**0.5)  / (2* self.nb[0])
                Sb = (self.nb[1] - delta**0.5)  / (2* self.nb[0])
                print(Sa, Sb)
            
        else:
            return f"{self.nb[0]} = {- self.nb[1] + rep} est solution de l'équation"
        
        
    def __str__(self):
        
        affichage = []
        
        for i in range(len(self.nb) - 2):
            affichage.append((self.nb[i] * self.signe[i], "x**", (len(self.nb) - i - 1)))
        affichage.append(((self.nb[len(self.nb) - 2] * self.signe[len(self.nb) - i - 1]), "x"))
        affichage.append(self.nb[len(self.nb) - 1] * self.signe[len(self.nb)- i])

        for tuple in range (len(affichage)-1 ):
            for element in affichage[tuple]:
                print(element, end = '')
            print(end = ' + ')
        print(affichage[len(affichage)])
        
    def __repr__(self):
        affichage = []
        
        for i in range(len(self.nb) - 2):
            affichage.append((self.nb[i] * self.signe[i], "x**", (len(self.nb) - i - 1)))
        affichage.append(((self.nb[len(self.nb) - 2] * self.signe[len(self.nb) - i - 1]), "x"))
        affichage.append(self.nb[len(self.nb) - 1] * self.signe[len(self.nb)- i])
        
        
        for tuple in range (len(affichage)-1):
            for element in affichage[tuple]:
                print(element, end = '')
            print(end = ' + ')
        print(affichage[len(affichage)-1])
```

```python
Inconnu = Equation([-3, 8, -7, 0, 9])

print(Inconnu)
```
