```python
def odg(nb, n=0):
    """
    >>> odg(45278)
    4
    """
    
    nb = nb // 10
    if nb < 1:
        return n
    return odg(nb, n+1)
        
    
def n_palindrome(nb, ordre_grandeur=None):
    """
    >>> n_palindrome(133454331)
    True
    >>> n_palindrome(12300321)
    True
    >>> n_palindrome(4141)
    False
    """
    if ordre_grandeur == None:
        ordre_grandeur = odg(nb)
    elif ordre_grandeur < 2: return True
    
    if nb // 10 ** ordre_grandeur == nb % 10: return n_palindrome((nb % 10 ** ordre_grandeur) // 10, ordre_grandeur - 2)
    else: return False
    

if __name__ == "__main__":
    import doctest
    

    doctest.testmod()
```
